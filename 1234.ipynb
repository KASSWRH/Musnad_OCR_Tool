```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

#   Bounding Boxes
annotations = [
    {
        "bbox": {
            "x": 1031.0702942159733,
            "y": 590.0655210336622,
            "width": 115.86575189467499,
            "height": 168.00534024727892
        },
        "label": ""
    },
    {
        "bbox": {
            "x": 868.858241563428,
            "y": 578.4789458441948,
            "width": 136.14225847624334,
            "height": 139.03890227361012
        },
        "label": ""
    }
]

#  
fig, ax = plt.subplots(figsize=(10, 8))

#   
ax.set_facecolor('white')

#   Bounding Boxes
for annotation in annotations:
    bbox = annotation['bbox']
    x, y, w, h = bbox['x'], bbox['y'], bbox['width'], bbox['height']
    
    #   Bounding Box
    rect = patches.Rectangle((x, y), w, h, linewidth=3, edgecolor='red', facecolor='none')
    ax.add_patch(rect)
    
    #   (  )
    if annotation['label']:
        ax.text(x, y - 10, annotation['label'], color='red', fontsize=12, weight='bold')

#  
ax.set_xlim(0, 1500)
ax.set_ylim(0, 1109)
ax.set_aspect('equal')
ax.axis('off')  #  

#  
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

#   
image_path = "D:\\casi_images_all\\dataset\\4_high.jpg"

#  
image = Image.open(image_path)

#   Bounding Boxes
annotations = [
    {
        "bbox": {
            "x": 1031.0702942159733,
            "y": 590.0655210336622,
            "width": 115.86575189467499,
            "height": 168.00534024727892
        },
        "label": ""
    },
    {
        "bbox": {
            "x": 868.858241563428,
            "y": 578.4789458441948,
            "width": 136.14225847624334,
            "height": 139.03890227361012
        },
        "label": ""
    }
]

#  
fig, axes = plt.subplots(1, len(annotations), figsize=(12, 6))

#        axes  
if len(annotations) == 1:
    axes = [axes]

#         Bounding Boxes
for ax, annotation in zip(axes, annotations):
    bbox = annotation['bbox']
    x, y, w, h = bbox['x'], bbox['y'], bbox['width'], bbox['height']
    
    #   Bounding Box   
    cropped_image = image.crop((x, y, x + w, y + h))
    
    #   
    ax.imshow(cropped_image)
    ax.set_title(f"Label: {annotation['label']}")
    ax.axis('off')  #  

#  
plt.tight_layout()
plt.show()
```

```python
import json
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import numpy as np

#  
image_path = r"data/project_test_fdb0a04a-f6e5-49f7-9e77-f3e1ed8c792b/processed_images/b93d7049-9b40-4fce-a75a-274d5168f1d7.jpg"
json_path  = r"data/project_test_fdb0a04a-f6e5-49f7-9e77-f3e1ed8c792b/annotations/b93d7049-9b40-4fce-a75a-274d5168f1d7.json"

#     JSON
with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

#   
image = cv2.imread(image_path)
if image is None:
    raise FileNotFoundError(f"       : {image_path}")

#   RGB      PIL
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_pil = Image.fromarray(image)
draw = ImageDraw.Draw(image_pil)

#      (    )
font_path = r"sf_old_south_arabian_serif.ttf"   #       
font = ImageFont.truetype(font_path, size=48)

#    
for ann in data.get("annotations", []):
    bbox = ann["bbox"]
    label = ann["label"]
    x, y, w, h = bbox["x"], bbox["y"], bbox["width"], bbox["height"]

    #    
    draw.rectangle(
        [(x, y), (x + w, y + h)],
        outline=(255, 0, 0),
        width=4
    )

    #    
    draw.text(
        (x, y - 50),   #    
        label,
        font=font,
        fill=(0, 255, 0)
    )

#      numpy   matplotlib
image_result = np.array(image_pil)

#    
plt.figure(figsize=(10, 10))
plt.imshow(image_result)
plt.axis("off")
plt.title("Bounding Boxes and Musnad Labels (  )", fontsize=14)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

#  
image_path = r"exports/yolo_fdb0a04a-f6e5-49f7-9e77-f3e1ed8c792b_20251016_194930/images\test/5_high.jpg"
label_path = r"exports/yolo_fdb0a04a-f6e5-49f7-9e77-f3e1ed8c792b_20251016_194930/labels/test/5_high.txt"

#    (      )
#   (     )
class_names = {
    0: "",
    1: "",
    2: "",
    3: ""
}

#   
image = cv2.imread(image_path)
if image is None:
    raise FileNotFoundError(f"      : {image_path}")

#  BGR  RGB  
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
height, width, _ = image.shape

#    
with open(label_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

#    bounding boxes
for line in lines:
    cls, x_center, y_center, w, h = map(float, line.strip().split())

    #      
    x_center *= width
    y_center *= height
    w *= width
    h *= height

    #   
    x1 = int(x_center - w / 2)
    y1 = int(y_center - h / 2)
    x2 = int(x_center + w / 2)
    y2 = int(y_center + h / 2)

    #  
    cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 3)

    #     
    label_name = class_names.get(int(cls), str(int(cls)))
    cv2.putText(image, label_name, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                1.2, (0, 255, 0), 3, cv2.LINE_AA)

#   
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis("off")
plt.title("YOLO Predictions ( )", fontsize=14)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import numpy as np

#  
image_path = r"exports/yolo_fdb0a04a-f6e5-49f7-9e77-f3e1ed8c792b_20251016_194930/images\test/5_high.jpg"
label_path = r"exports/yolo_fdb0a04a-f6e5-49f7-9e77-f3e1ed8c792b_20251016_194930/labels/test/5_high.txt"

#    (         )
class_names = {
    0: "",
    1: "",
    2: "",
    3: ""
}

#   
image = cv2.imread(image_path)
if image is None:
    raise FileNotFoundError(f"      : {image_path}")

#   RGB    Pillow
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
height, width, _ = image.shape
image_pil = Image.fromarray(image)
draw = ImageDraw.Draw(image_pil)

#      
font_path = r"sf_old_south_arabian_serif.ttf"  #         
font = ImageFont.truetype(font_path, size=48)

#     (YOLO TXT)
with open(label_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

#    bounding boxes   
for line in lines:
    cls, x_center, y_center, w, h = map(float, line.strip().split())

    #     
    x_center *= width
    y_center *= height
    w *= width
    h *= height

    #  
    x1 = int(x_center - w / 2)
    y1 = int(y_center - h / 2)
    x2 = int(x_center + w / 2)
    y2 = int(y_center + h / 2)

    #    
    draw.rectangle([(x1, y1), (x2, y2)], outline=(255, 0, 0), width=4)

    #     (  ... )
    label_name = class_names.get(int(cls), str(int(cls)))
    draw.text((x1, y1 - 55), label_name, font=font, fill=(0, 255, 0))

#      NumPy   matplotlib
image_result = np.array(image_pil)

#    
plt.figure(figsize=(10, 10))
plt.imshow(image_result)
plt.axis("off")
plt.title("YOLO Predictions   ", fontsize=14)
plt.show()
```
